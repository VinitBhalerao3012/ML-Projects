```python
import pandas as pd
```

```python
df = pd.read_csv("seattle-weather.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
def LabelEncoding(c):
    from sklearn import preprocessing
    le = preprocessing.LabelEncoder()
    df[c]= le.fit_transform(df[c])
    df[c].unique()
LabelEncoding("weather")
df
```

```python
cols = ['precipitation' , 'temp_max', 'temp_min', 'wind']
```

```python
def normalize(df,cols):
    for x in cols:
        df[x] = df[x]/df[x].max()
normalize(df,cols)
df
```

```python
df = df.drop('date',axis=1)
df
```

```python
x = df.drop('weather',axis=1)
y = df['weather']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from xgboost import XGBClassifier
print("XGBoost works")
```

```python
xg = XGBClassifier()
xg.fit(X_train, y_train)
```

```python
xg.get_params()
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
y_hat = xg.predict(X_test)
print("Accuracy Score =",accuracy_score(y_test,y_hat))
print("Classification Report")
print(classification_report(y_test,y_hat))
```

```python
grid = {'learning_rate': [0.1,1, 0.01, 0.001], 'gamma':[0,1,10,100]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(XGBClassifier(), grid, cv=10, verbose=2)
```

```python
model.fit(X_train, y_train)
```

```python
grid_predictions = model.predict(X_test)
print("Accuracy Score = ",accuracy_score(y_test,grid_predictions))
print("Classification Report")
print(classification_report(y_test,grid_predictions))
```

```python
print(model.best_estimator_)
```
